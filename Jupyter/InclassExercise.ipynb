## Overview of Jupyter Notebooks
A Jupyter Notebook is an interactive environment that allows you to combine: 
* Live code (in Python, R, Julia and more)
* Text and formatting (using Markdown)
* Visualizations (e.g.,plots, charts)
* Interactive widgets

All of this is organized into cells, which can be: 
1. Code cells - to write and run code.
2. Markdown cells - for explanatory text, notes, or documentation.
