## Variables and functions

So far, we have only used Python in such a way that we give it a bunch of literals, evaluate an *expression*, and get a bunch of literals. But this is not much more useful than a fancy (and clunky!) calculator.

Python allows us to associate an object (think "some value", for now) with a name of our choosing. 

For example:

````python
x = 42
````

After you run the cell above (did you run it?), everytime you type `x` python will interpret it as `42`.

````python
x
````

````python
x + 2
````

````python
y = x + 2
y
````

Notice how we created a new name above called `y`. But it is also important to note that what we did above is **not** the same as this:
$$
y = x + 2
$$

That is because the above is a **function** often represented as $f(x) = x + 2$. In plain English it says, "take a given value of $x$, and return 2 more than that value."

Where as this Python code:
```python
y = x + 2
```
In plain English, says the following:

> Evaluate the expression on the right of the `=` and assign the result to the name (variable) on the left of the assignment operator(`=`).

In fact, in Python (and most programming languages) we call the equal sign, the *assignment operator* **because** it assigns values to variables.

**Prove it**

Check your understanding by changing the value of `x` and determining if the value of `y` correspondingly changes.

````python
# write code to change the value of x
x = 32 
y = x 
y # check if y changed too
````

Okay so how would we express a **function** in Python. We can also assign a function to a name, but there is a special way to do it:

````python
def f(x): # note: the name of the function can be any name we want!
    return x + 2
````

Now whenever we enter `f(of_some_value)` Python will return (replace that) with the result of the function.

````python
f(42)
````

### Variables can store multiple values

Python variables are not restricted to holding just one value. They can hold **multiple** values: 

```python
X = [10, 20, 30, 40]
```

Notice:
- The square brackets - they indicate to Python that this is a list.
- The commas that seperate individual values.

The best way to think of a list, is as though we are referencing a column in Excel. We can reference the whole column or we can reference individual values.

To access individual values in sequences such as lists, we use a special notation called *subscript*. For example, to get the second value in a list we do the following:
```python
x[1]
```
⚠️ **Note**: That wasn't a typo. In programming, we often start numbering at zero!

````python
from sympy import symbols, plot #hint if you get module sympy not found, use the technique above to install it
````

````python
x = symbols('x', integer=True, positive=True)
print(x) 
print(x.is_positive) #True
print(x.is_integer) #True 
````

````python
x = symbols('x')


#f = 2*x + 1
f = x ** 2 + 2 * x + 1 # Quad
plot(f, (x, -20, 20), title = "Test plot with sympy", line_color='g', marker='ob');
````
