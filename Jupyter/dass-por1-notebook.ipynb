#### Dictionaries

Dictionaries are another useful data structure that allow us to look up values by a key. Like lists, dictionaries allow us to store **any** values but the keys are limited to data types that never change (for now, think strings or ints). Here is an example of how to create and access a dictionary's value:

```python
week_days = {
    'Sunday': 1,
    'Monday': 2,
    'Tuesday': 3,
    'Wednesday': 4,
    'Thursday': 5,
    'Friday': 6,
    'Saturday': 7
}
# Accessing the value associated with Tuesday
tuesday_value = week_days['Tuesday']

# Printing the value
print("Numeric representation of Tuesday:", tuesday_value)
```
Create a dictionary named `prices` with keys as the product names and values as their respective prices: Apples - 1.5, Oranges - 2.0, Bananas - 0.5. After you have created the dictionary, retrieve the price of oranges and display that as the result of the evaluation of the cell.

````python
prices = {
    'Apples' : 1.5,
    'Oranges' : 2.0,
    'Bananas' : 0.5
} 
# 
price_range_orange = prices['Oranges']
# 
print(f"The price range of oranges : ${price_range_orange}")
````

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(42)

# Generate random data
x_data = np.arange(0, 100)  # An array of 0 to 99
y_data = x_data * 0.5 + np.random.normal(size=x_data.size) * 10  # Linear relation with noise
# Creating a scatter plot with a regression line
sns.set_theme(style="whitegrid")  # Set theme for aesthetics
plt.figure(figsize=(10, 6))  # Set figure size
sns.regplot(x=x_data, y=y_data, scatter_kws={"color": "blue"}, line_kws={"color": "red"})

# Adding title and labels
plt.title('Random Data Visualization with Regression Line')
plt.xlabel('X Data')
plt.ylabel('Y Data')

# Show the plot
plt.show()
````
