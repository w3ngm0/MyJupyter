## KMeans Clustering Analysis

### Task Determine the Optimal Number of Clusters

The code below calculates and plots inertia for various cluster sizes for KMeans. But there is a small bug in the code that you will need to fix before it will run successfully!

````python
from sklearn.cluster import KMeans

k_values = range(1, 10)
inertia = []

features = df_scaled.iloc[:,:-1]
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(8, 4))
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.xticks(k_values)
plt.show()
````

### Task: Examine cluster separation

Another measure that is commonly used with K-Means clustering is the silhouette score. The silhouette score identifies how close points are to the peers in their respective clusters compared to points in other clusters. It provides a score between -1 and 1. A score of 1 suggests clearly defined boundaries between clusters (points within a cluster are grouped closer together, and each cluster is far from other clusters). Negative scores suggest misclassifying points to clusters (and may often indicate a process error rather than a data set feature).
> **Note**: You have to fix the bug in the previous code for this to run!


````python
from sklearn.metrics import silhouette_score

k_values = range(2, 7)
silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(features)
    silhouette_avg = silhouette_score(features, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot the Silhouette Scores
plt.figure(figsize=(8, 4))
plt.plot(k_values, silhouette_scores, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score For Different k')
plt.xticks(k_values)
plt.show()
````

#### Generate a Confusion Matrix

One of the most useful visualizations we can create to evaluate a classification model is a confusion matrix. Supposedly, the term confusion refers to "the confusion of the model" but it may also be that they can be a bit confusing at first.

A confusion matrix provides a 2D plot of actual and predicted values, allowing us to determine both **false positive** and **true positive** as well as **false negatives** and **true negatives**, depending on our business case we may put more weight on any of these. Where confusion matrices can get confusing is that we often want to determine these values where there are more than 2 classifications.

Further, to create a confusion matrix, we need to compare like-for-like and recall that in our example we started with a species (string) and what we got were labels (integers between 0-3). So if we try and see how often our clusters (y_pred) match the species, we are going to find "never" as the answer.

So, in order to even begin a comparison, we need to *map* the clusters to their respective names or what they mean to us - but how?

A common way is to start with optimism and assume that the guess the model makes most frequently is probably right and then we have a mapping. Let's look at an example from our data set:  


# Scaled Matrix

### Task: Formulate a Hypothesis

The mean value between the two groups is unquestionably different. It's just a fact. But this fact only applies to our particular sample. We want to try to figure out how likely it is that this difference will be found in the population. That is we want to test its significance.

Significance testing and hypothesis testing are sometimes used interchangeably.

Given the species and feature you found, formulate a hypothesis:

> Given the two species for the iris dataset, my null hypothesis is  that there is significant difference between the mean average sepal length of setosa and virginica.
The alternative hypothesis is that there is a significant difference between the mean sepal lenght of setosa and virginica. 

To develop confidence in our hypothesis, we want to ask the following question:

*Suppose that in our population, on average, there is no difference between the two populations. How likely would we find a difference like the one we did by chance alone?*

The hypothesis that chance alone can explain a difference between groups is called the **null hypothesis**.

The alternative is that it is doubtful that we would get the difference we measured by chance alone, which is called the **alternative hypothesis**.
[Reference](https://learning.oreilly.com/library/view/practical-statistics-for/9781492072935/ch03.html#idm45782046480184)

We usually start with the null hypothesis and try to prove how likely it is to be true; if the chance that the null hypothesis is true is very low (different contexts set different thresholds by convention), then we reject the null hypothesis. The probability of getting the difference we observed assuming the populations are identical concerning that feature is typically called the *p-value*. In the business and humanities, a *p-value* of less than 0.05 (meaning less than a 5% chance of getting the difference we observed by chance) is said to be statistically significant.

A common test we can perform on samples is called the t-test and we can use the scipy module to perform advanced statistical analysis

````python
# Use a more meaningful name (that you came up with earlier)
from scipy.stats import ttest_ind
result = ttest_ind(versicolor_group, virginica_group, equal_var=False)
print(result)
````

> Since p-value 0.0018 < 0.05, I reject the H0(null hypothesis). The result indicates that there is a statistically significant difference between the mean sepal widths of virginica and versicolor.
