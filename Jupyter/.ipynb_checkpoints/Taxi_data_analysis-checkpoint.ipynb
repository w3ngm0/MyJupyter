# 📓 Project: NYC Taxi Tipping Analysis

**Objective**: Analyze tip rate by region (borough) in NYC taxis.

**Workflow**: Load → Explore → Clean (filter cash) → Feature Engineering → Aggregation (by region) → Visualization → Insights

```python
# Step 1: Load the data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV
df = pd.read_csv('taxi_analysis_project.csv')
print(f"Loaded {len(df):,} rows × {len(df.columns)} columns")
df.head()
```

## Step 2: Initial exploration

- Dataset shape and column types
- Missing values
- Basic statistics

```python
# Step 2: Initial exploration (shape, dtypes, missing values)
print("=== Dataset shape ===")
print(df.shape)

print("\n=== Column types ===")
print(df.dtypes)

print("\n=== Missing values ===")
print(df.isnull().sum())

print("\n=== Basic statistics ===")
df.describe()
```

## Step 3: Data cleaning

- Parse datetime columns
- Handle missing values
- Remove invalid rows (distance/fare <= 0)
- **Filter out cash payments** (cash tips are usually not recorded)

```python
# Parse datetime columns
df['pickup'] = pd.to_datetime(df['pickup'])
df['dropoff'] = pd.to_datetime(df['dropoff'])

# Check for invalid numeric values (zeros, negatives where impossible)
print("Rows with distance <= 0:", (df['distance'] <= 0).sum())
print("Rows with fare <= 0:", (df['fare'] <= 0).sum())
print("Rows with total <= 0:", (df['total'] <= 0).sum())
print("Rows with passengers < 0:", (df['passengers'] < 0).sum())
```

```python
# Drop rows with invalid numeric data (distance or fare <= 0)
df_clean = df[(df['distance'] > 0) & (df['fare'] > 0)].copy()

# Filter out cash payments (cash tips are usually not recorded)
df_clean = df_clean[df_clean['payment'] != 'cash'].copy()

# Fill missing categorical values
df_clean['payment'] = df_clean['payment'].fillna('unknown')
df_clean['pickup_borough'] = df_clean['pickup_borough'].fillna('Unknown')
df_clean['dropoff_borough'] = df_clean['dropoff_borough'].fillna('Unknown')

print(f"Rows after cleaning: {len(df_clean):,} (removed {len(df) - len(df_clean):,})")
df_clean.head()
```

## Step 4: Feature engineering (for tip rate by region)

- Tip rate (tip / fare)
- Region = pickup_borough

```python
# Trip duration in minutes
df_clean['duration_min'] = (df_clean['dropoff'] - df_clean['pickup']).dt.total_seconds() / 60

# Tip rate: tip as % of fare (for credit-card trips where tip is recorded)
df_clean['tip_rate'] = np.where(df_clean['fare'] > 0, df_clean['tip'] / df_clean['fare'], 0)

# Time features
df_clean['hour'] = df_clean['pickup'].dt.hour
df_clean['day_of_week'] = df_clean['pickup'].dt.day_name()
df_clean['is_weekend'] = df_clean['pickup'].dt.dayofweek >= 5

# Cross-borough trip
df_clean['cross_borough'] = df_clean['pickup_borough'] != df_clean['dropoff_borough']

# Fare per mile
df_clean['fare_per_mile'] = df_clean['fare'] / df_clean['distance']

df_clean[['pickup_borough', 'tip', 'tip_rate', 'fare']].head(10)
```

## Step 5: Aggregation – tip rate by region

- Avg tip rate by pickup borough
- Avg tip rate by pickup zone (more granular)

```python
# Tip rate by region: aggregate by pickup borough
by_region = df_clean.groupby('pickup_borough').agg(
    count=('tip', 'count'),
    avg_tip=('tip', 'mean'),
    avg_tip_rate=('tip_rate', 'mean'),
    avg_fare=('fare', 'mean')
).round(4)
by_region['avg_tip_rate_pct'] = (by_region['avg_tip_rate'] * 100).round(1).astype(str) + '%'
by_region
```

```python
# Tip rate by pickup zone (more granular)
by_zone = df_clean.groupby('pickup_zone').agg(
    count=('tip', 'count'),
    avg_tip_rate=('tip_rate', 'mean')
).round(4).sort_values('avg_tip_rate', ascending=False)
by_zone = by_zone[by_zone['count'] >= 20]  # filter zones with enough trips
by_zone.head(15)
```

```python
# Drop Unknown borough for cleaner regional analysis
by_region_clean = by_region[by_region.index != 'Unknown']
by_region_clean
```

## Step 6: Visualization

- Tip rate by region (borough)
- Tip rate by zone

```python
sns.set_style("whitegrid")
fig, axes = plt.subplots(2, 1, figsize=(10, 10))

# Tip rate by borough
by_plot = by_region[by_region.index != 'Unknown'].sort_values('avg_tip_rate', ascending=True)
axes[0].barh(by_plot.index, by_plot['avg_tip_rate'] * 100)
axes[0].set_xlabel('Avg tip rate (%)')
axes[0].set_title('Tip rate by pickup borough')
axes[0].set_xlim(0, None)

# Tip rate by zone (top 15)
top_zones = by_zone.head(15)
axes[1].barh(range(len(top_zones)), top_zones['avg_tip_rate'] * 100)
axes[1].set_yticks(range(len(top_zones)))
axes[1].set_yticklabels(top_zones.index, fontsize=9)
axes[1].set_xlabel('Avg tip rate (%)')
axes[1].set_title('Tip rate by pickup zone (top 15, min 20 trips)')
axes[1].set_xlim(0, None)
plt.tight_layout()
plt.show()
```

```python
# Box plot: tip rate by borough
plt.figure(figsize=(8, 5))
df_region = df_clean[df_clean['pickup_borough'] != 'Unknown']
sns.boxplot(data=df_region, x='pickup_borough', y='tip_rate', order=by_region_clean.sort_values('avg_tip_rate').index)
plt.xticks(rotation=45)
plt.ylabel('Tip rate')
plt.xlabel('Pickup borough')
plt.title('Tip rate distribution by region')
plt.tight_layout()
plt.show()
```

## Step 7: Insights & conclusion

- Tip rate varies by region
- Borough and zone-level patterns

```python
print("=== Key insights ===\n")
print(f"1. Dataset: {len(df_clean):,} trips (cash excluded; tips not recorded for cash)")
print(f"2. Overall avg tip rate: {df_clean['tip_rate'].mean()*100:.1f}%")
print(f"3. Tip rate varies by region (see borough and zone tables above)")
print(f"4. Manhattan dominates trip volume; compare tip rates across boroughs for regional patterns")
```
