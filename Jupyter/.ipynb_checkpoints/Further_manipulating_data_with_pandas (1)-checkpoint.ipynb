### Series

- A Series is a one-dimensional array-like object containing an array of data and an associated array of data labels called its index.
- Reflecting on the spreadsheet analogy, we can think of a Series as a single spreadsheet column.
- In addition to the column data, a series contains an index
- A series can be create from a list-like object
- Technical: one-dimensional array
  
### Dataframes 

- DataFrames themselves composed of Series objects.
- Even though we typically work with DataFrames, it is crucial to understand Series as they are the building blocks of DataFrames.
  
#### So, what's the point of using a Series?

- **Structurally:**
    - Unlike a list, you have a positional and labelled index. This means we can access elements using the index.
    - you can use a dictionary as a mapping to a Series with a labelled index

- **Implementation:**
    - Built on NumPy
    - Vectorized
    - Optimized


````python
import pandas as pd
````

````python
# Example: Creating a series from a list

data = pd.Series([0.25, 0.5, 0.75, 1.0])
print(data)

# recall that you can access elements of a list using the index. 
print(data[1]) # produces 0.5
````


 
In the first example we created, we didn't specify an index, so the index isn't all that different from a list index. However, we can specify an index when we create a Series. For example:

```python

```

Now, we can access elements using the index:


```python

```


Alternatively, suppose we want to add an index to an existing Series:


```python

```

Okay, but isn't that just a dictionary? 

Yes, it is. In fact, you can create a Series from a dictionary. For example:


```python

                   'Florida': 19552860,
                   'Illinois': 12882135}

```

All you have shown is that a Series is like a list and a dictionary. What's the point? 

The point is that a Series is like a list and a dictionary, but it has additional functionality that makes it useful for data analysis. For example, we can perform vectorized operations on a Series, which we can't do with a dictionary and we can store primitive data types, which we can't do with any built-in Python object. 

Let's deal with each of these concepts in turn.

## Vectorized operations

- Python for loops are too slow for large-scale data analytics
- For loops are executed serially (one after the other)
- Vectorized operations are carried out on all elements at the same time
- In practice it means that operations on a series are applied to all elements in the series

## More about Series and DataFrames

### 1. Vectorized operations 

In Python, when we want to perform an operation on every element in a list, we typically use a for loop. 

For example:

```python
data = [0.25, 0.5, 0.75, 1.0]
result = []
for element in data:
    result.append(element * 2)
result
```

This is a pefectly fine way of doing things for small lists, but it is slow for large lists. 

Compared to programming languages that data scientists traditionally used, like C or FORTRAN, it is snail paced. Not just frustratingly slow, but impractically slow for some tasks.

But let's consider the serial nature of the for loop: we process items one at a time, but in the example we used we just wanted to double each element. Since the result of the previous operation has no impact on the next, we could have (at least in theory) done it all at once. 

Vectorized operations are operations that are applied to every element in a data structure at once. This is much faster than using a for loop.

Usually the vectorized operations are implemented in a low-level language like C, so they are much faster than using a for loop in Python.

> Note: The term "vectorized" comes from the fact that the operations are applied to every element in a vector at once. In linear algebra, a vector is a one-dimensional array, which is what a Series is.

The actual implementation of vectorized operations is handled by NumPy, which is a library that is used by Pandas. We will cover NumPy in a later lesson.

For now, let's just see how we can use vectorized operations with a Series. For example:
```python
data = pd.Series([0.25, 0.5, 0.75, 1.0])
result = data * 2
result
```
Not only is this faster than using a for loop, but it is also more concise and easier to read.

````python
data = pd.Series([0.25, 0.5, 0.75, 1.0])
result = data * 2
result
````

### 2. Comparison and Boolean operations 
We can also perform comparison operations on a Series. 

Example:

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0])
data > 0.5
```

Try running the previous code, and then check what the type is of the output (`type(data > 0.5)`)


````python
# TODO: Copy above code and check the type of the output
data = pd.Series([0.25, 0.5, 0.75, 1.0])
#data > 0.5
(data > 0.5) & (data < 0.8)
(data > 0.5) | (data < 0.8)
````

Any comparison operation we perform on a Series will return a Series of booleans.

All the comparison operators we learned about can be used on a series and combined using logical operators: `&` (and), `|` (or), and `~` (not). 

For example:
```python
(data > 0.5) & (data < 0.8)
# or
(data > 0.5) | (data < 0.8)
```

Notice that we use parentheses to group the boolean operations. This is because the `&` and `|` operators have higher precedence than the comparison operators.
Notice also that we use the `&` and `|` operators instead of `and` and `or`. This is because the `and` and `or` and for that matter 'not' operators cannot be repurposed for vectorized operations. So pandas had to use the `&` and `|` and `~` operators that traditionally represent bitwise operations to do vectorized boolean operations.

### 3. Filtering series

The most common use of comparison and boolean operations is to filter lists. This takes advantage of a property of a series subscript that it can take a series of booleans and use them as a mask. 

The most common use of boolean operations on a Series is to filter data. For example:

```python
data[data > 0.5]
```

This will output:

```
2    0.75
3    1.00
dtype: float64
```

We can think of the boolean operation as a mask that we apply to the Series. The mask is a Series of booleans, and the Series that we apply the mask to is the same length as the mask. The mask is used to select the elements of the Series that we want to keep.

The following series shows the population of each state in Australia:

````python
population = pd.Series({'Queensland': 5110000,  # notice that we are making a series from a dictionary here
                   'New South Wales': 8089526,
                   'Victoria': 6680644,
                   'Tasmania': 541071,
                   'South Australia': 1751693,
                   'Western Australia': 2621680,
                   'Northern Territory': 245869,
                   'Australian Capital Territory': 390800})

population.index.to_list() # displays just the index as a list

print(f"""

All the states in the series:

{', '.join(population.index.to_list())}

""")
````
