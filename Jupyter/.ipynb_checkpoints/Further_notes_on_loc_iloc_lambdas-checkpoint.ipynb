### Let's investigate how loc[] and iloc[] attributes work in DataFrames

For this, let's use the **tips** dataset which is about tipping details at a particular restaurant.

````python
import pandas as pd
import seaborn as sns
````

````python
df_tips = sns.load_dataset("tips")
````

````python
print( df_tips.info())
print( df_tips.head())
````

**Creating a DataFrameGroupBy object to calculate group statistics**

````python
dfg_tips_day = df_tips.groupby('day', observed = True)

# Note the data type of 'dfg_tips_day' variable
print( type(dfg_tips_day))
````

````python
# Use the 'DataFrameGroupBy' object for summary stats
print( "Average tip value (by day) = ", dfg_tips_day['tip'].mean())
print( "Total count of tips recorded (by day) = ", dfg_tips_day['tip'].count())
````

### **using ***loc*** attribute to access values**

- range of rows/columns can be specified using : operator
- if multiple non-consecutive rows/columns are specified, use a list

````python
# accessing value at row index 1, column 'time'

df_tips.loc[1,"time"]
````

````python
# accessing values at row index 1, columns 'time', 'sex'

df_tips.loc[1,["time", "sex"]]
````

````python
# accessing values at multiple rows and columns

df_tips.loc[[1,3,4],["time", "sex"]]
````

````python
# specifying a range of rows when using loc

df_tips.loc[2:5,["time", "sex"]]
````

### **using ***iloc*** attribute to access values**

- range of rows/columns can be specified using : operator
- if multiple non-consecutive rows/columns are specified, use a list
- Note that we cannot use column names with iloc

````python
df_tips.iloc[2:5,[0,2]]
````

````python
df_tips.iloc[10:20,:3]
````

**using loc to modify filtered rows**

````python

#create a new column named 'level
df_tips['level'] = "average"

# set 'level' of rows with tips higher than $5 as 'high'
df_tips.loc[ df_tips['tip']>5, 'level'] = 'high'

# set 'level' of rows with tips lower than $1.5 as 'low'
df_tips.loc[ df_tips['tip']>5, 'level'] = 'low'

````

## 4. Lambdas and why we should care

Lambdas are python anonymous functions. They are usually a more advanced topic and often are a sign of poor maintainability in code.

However, in the context of vectorized processing they can be really useful as they allow us to `apply` to a series a function that is more complex than a simple arithmatic operation.

Let's look at some examples of lambdas and then apply them to a series.

Here's a regular Python function that takes a name and returns "Hello, <name>":

```python
def greet(name):
    return f"Hello, {name}"
```

Here's the same function as a lambda:

```python
greet = lambda name: f"Hello, {name}"
```

Here's a slightly more complex example that takes a temperature in Celsius and returns the temperature in Fahrenheit.

First, as a function:

```python
def celsius_to_fahrenheit(celsius):
    return celsius * 9/5 + 32
```

Now, as a lambda:

```python
celsius_to_fahrenheit = lambda celsius: celsius * 9/5 + 32
```

Now, let's imagine that we have a series of average March temperatures in various capitals in Australia in Celsius:

```python
march_temperatures = pd.Series({'Sydney': 23.5,
                                'Melbourne': 20.5,
                                'Brisbane': 25.3,
                                'Perth': 24.7,
                                'Adelaide': 22.3,
                                'Hobart': 18.7,
                                'Darwin': 31.8,
                                'Canberra': 20.9})
```

We can use the `apply` method to apply a function to a series. Notice that, as written, it makes no difference whether the function is a regular function or a lambda. For example:

```python
march_temperatures.apply(celsius_to_fahrenheit)
```

But where the lambda has an advantage is that it allows us to make the definition explicit in the context where it is used, like so:

```python
march_temperatures.apply(lambda c: c * 9/5 + 32)
```


### Example: Lambda functions vs normal functions

````python
# just a normal function
def addTen( a):
    return a + 10
````

````python
#lambda eqvivalent to ablve function
lambda a : a + 10
````

````python
# Using the normal function
addTen( 35)
````

````python
# just a normal function
def calcDiscountedPrice( full_price, discount_rate, coupon_value):
    return full_price * ( 1 - discount_rate) - coupon_value
    
````

````python
#lambda eqvivalent to above function
x = lambda full_price, discount_rate, coupon_value: full_price * ( 1 - discount_rate) - coupon_value
````

````python
# Using the normal function
calcDiscountedPrice( 800, 0.05, 20)
````

````python
calcDiscountedPrice( 1000, 0.01, 50)
````

````python
# Using the lambda functions
x( 800, 0.05, 20)
````

### using ***apply( )*** function on a series

Note that the input parameter to apply( ) is a *reference to another function*.

````python
# create a series that holds temperature values for the month of March in Austrlian capitals
march_temperatures = pd.Series({'Sydney': 23.5,
                                'Melbourne': 20.5,
                                'Brisbane': 25.3,
                                'Perth': 24.7,
                                'Adelaide': 22.3,
                                'Hobart': 18.7,
                                'Darwin': 31.8,
                                'Canberra': 20.9})
````

````python
# note what the index holds
march_temperatures.index
````

````python
# normal function to convert celcius to fahrenheit
def celsius_to_fahrenheit(celsius):
    return celsius * 9/5 + 32
````

````python
# calling the above function
celsius_to_fahrenheit(march_temperatures)
````

````python
# using a lambda function in apply() to convert celcius to fahrenheit
march_temperatures.apply( lambda cel: cel * 9/5 + 132)
````
