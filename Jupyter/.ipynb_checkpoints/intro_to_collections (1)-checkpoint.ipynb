### Categorizing Collections
Collections can be categorized into mutable or umutable; ordered or unordered:
- **Mutable**: can be modified
- **Ordered**: can be accessed by position, order matters

| Data Type | Mutable | Ordered |
| :--- | :---: | :---: |
| `list` | ✅ | ✅ |
| `tuple` | ❌ | ✅ |
| `dict` | ✅ | ❌ |
| `set` | ✅ | ❌ |

```python
{'a': 1, 'b': 2} == {'b': 2, 'a': 1}
```

### Ordered

An ordered collection can be accessed by an index. The index uses subscript notation meaning: `some_ordered_collection[pos]` where `pos` is an index from `0` to (`len - 1`).

The built in ordered collections are:
- strings
- lists
- tuples

We can also index from the right by using negative numbers!

#### Slicing
An ordered collection can also be sliced!

A slice of a collection takes the form of `some_ordered_collection[start:end:step]`

Where it is **inclusive** of the first value and **exclusive** of the last. The step is optional.

```python
x = "Monty, Python!"
# x = ["Monty", "Python", [1,2,3]]
```

```python
a = [5, 3, "hello", "x"]
b = [5, 3, "hello", "x"][::-1]
```

```python
a
```

```python
b
```

### Accessing dictionaries (unordered)

Values of dictionaries are also accessed with the subscript notation, but we use the key instead: `some_dictionary[key]` returns a value associated with a key.

You cannot access values positionally.

```python
key_values = {'key1': [1,2,3], 'key2': 'value2'} 
```

```python
key_values["key1"][1]
```
